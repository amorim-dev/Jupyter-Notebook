## 1. Open Jupyter Notebook

Open your terminal and enter in your directory using the following command:
> `cd folder-name/directory-name`

> `jupyter notebook`

<br>

![opening_jupyter](images/opening_jupyter.gif 'Opening jupyter notebook')

If your commands were successful, a new tab will open in your browser with the contents of the directory. You have to keep the terminal open.

<br>

## 2. Create a New File or Folder

To create a Python 3 jupyter file (.ipynb) or a folder, click on `New` and select what you want to create (eg. `Python 3`, `Folder`).


<br>

## 3. Rename a File or Folder

- To rename a folder only select the folder that you want to rename and click on rename.




- To rename a file, you have to open the file by clicking two times above the name, and then with a single click on the name of the file will open a pop-up to change the name.



<br>

## 4. Close Your Jupyter Notebook Session

- To close your jupyter files, select those files and click on `Shutdown`.



- After Shutdown you can end your jupyter Notebook session by clicking on `Quit`. After that, you can close the browser tab and the terminal.



**Note:** You can also shutdown a Jupyter Notebook session by clicking `Ctrl+C` in the terminal. You will be asked to confirm that you want to `Shutdown this notebook server (y/n)?`. Type `y + Enter` to confirm. Then, you can close the Terminal by typing the command `exit + Enter`.

<br>

## 5. Use Shortcuts to Increase Productivity

To use the shortcuts it's important to know that jupyter has to modes, and some of the shortcuts work only in a specific mode:

**Edit mode:** 
- We activate pressing `ENTER`
- Is represented with the **green color** on the left edge
- In this mode we can write our code and execute it.
![edit_mode.png](images/edit_mode.png)

**Command mode:** 
- We activate pressing `ESC` 
- Is represented with the **blue color** on the left edge
- In this mode we can make changes in the cell, such as delete cells, add new cells or change from markdown to python for example.
![command_mode.png](images/command_mode.png)


Here I will show some important shorcuts that you should know, but to see all the jupyter shortcuts press the letter `H` in the command mode and a pop up will appear to you. Try it for yourself.

**Important shortcuts that you should start to use today:**

**Work in EDIT MODE or COMMAND MODE**

Run code cells
- `Ctrl + Enter`: run selected cell
- `Alt + Enter`: run cell and insert another cell below
- `Shift + Enter`: run cell, select below or create a new one if is not any cell above

**Inside COMMAND MODE**

Insert, delete, copy or paste cells:
- `A`: insert a new cell **A**bove
- `B`: insert a new cell **B**elow
- `C`: copy selected cell
- `X`: cut selected cell
- `Shift + V`: paste cells above
 
Move into cells: 
(You can use the up and down arrows too)
- `J`: go to the cell below
- `K`: go to the cell above 

Replace values:
- `F`: find and replace
 
Change the cell mode:
- `M`: change cell to markdown
- `Y`: change cell to code

**Inside EDIT MODE**
- `Ctrl + A`: select all in the cell
- `Ctrl + D`: delete whole line
- `Ctrl + S`: Save and Checkpoint
- `Ctrl + Z`: undo
- `Ctrl + /`: comment

- `Tab`: code completion (if you start writing some code or variable) or indent.
- `Shift + Tab`: tooltip (show method documentation)
![](images/showMethodDoc.png 'To show method documentation')

## 6. Use Multi-cursor to edit multiple lines at the same time

- Hold down the `ctrl` key (`cmd` in Mac) and select all cells that you want to edit. 
- If you type in some place that you don't want to edit, you can delete the cursor by holding down `ctrl` and clicking again on the same place. 
- You will see that there are now multiple cursors and you can edit all the lines at the same time. 
- Use the arrows to move the cursor. 
- Once you are done editing, click with mouse to stop multiple cursor.

![mult_cursor](images/mult_cursor.gif 'Multicursor to edit multiple lines')

**Note:** You can also hold `alt + left-click` and drag your cursor.

<br>

## 7. A quickly way to add Pictures

- Change the cell type to `Markdown` with `M` in the command mode and just drag and drop the picture into the cell.
- Once youâ€™ve dropped the image into the cell, some code should appear. Run the cell `Shift + Enter` to see the image.

## 9. Use Built-in Magic Commands

**You can use the percent symbol `%` before some commands that you are used to use in the terminal as:**
> `%cd` - To change the current work directory

> `%ls` - To list the contents of the current directory

> `%history` - To show the history from the code (In [ ]:) commands 

> `%pwd` - To print the path of the current working directory

**List all variables in this notebook with `%who`**

```python
name = 'Bianca'
subject = 'Jupyter Notebook'
month = 'October'
year = 2022

%who
```

```python
# You can also specify the type that you want to see
%who int
```

**Measure line Execution time with `%timeit`or Cell Execution Time with `%%time`**

```python
# Measure line execution time
import numpy as np
array_python = range(20000)
%timeit new_array = [array_python[i] ** 4 for i in array_python]
```

```python
%%time
array_python = range(20000)
new_array = [array_python[i] ** 4 for i in array_python]
```

**Note:** `%%time` to measure the cell execution time has to be the firt code in your cell.

**Use `%lsmagic`to see the full list of magic commands:**

```python
%lsmagic
```

## 10. Hide Unnecessary Matplotlib Text

Use a semicolon `;` at the end of your code, to hide annoying and unnecessary text before your plot.

```python
from matplotlib import pyplot as plt
x = np.random.rand(30)
plt.hist(x)
```

```python
x = np.random.rand(30)
plt.hist(x);
```

## 12. Expand the output display to see more columns of a Pandas Dataframe

In Pandas, you have a limited number of rows and columns that are shown in a table, but you can customize using the method `set_option`. If you want more information you can find it in the pandas documentation at the following link: 

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.set_option.html

```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```
